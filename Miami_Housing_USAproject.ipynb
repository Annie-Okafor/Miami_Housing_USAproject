```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
miami_USA_project=pd.read_csv("/content/sample_data/miami-housing-USAproject.csv")
miami_USA_project
```

```python
miami_USA_project.drop(['LATITUDE', 'LONGITUDE','PARCELNO' ,'avno60plus', 'WATER_DIST'], axis = 1, inplace = True)
miami_USA_project
```

```python
M_U_p = miami_USA_project.reindex(['LND_SQFOOT','TOT_LVG_AREA','SPEC_FEAT_VAL','RAIL_DIST','OCEAN_DIST','CNTR_DIST','SUBCNTR_DI','HWY_DIST','age','month_sold','structure_quality','SALE_PRC'] , axis = 1)
M_U_p
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(M_U_p,test_size=0.12)
USA_project_xtrain=train.loc[:,['LND_SQFOOT','TOT_LVG_AREA','SPEC_FEAT_VAL','RAIL_DIST','OCEAN_DIST','CNTR_DIST','SUBCNTR_DI','HWY_DIST','age','month_sold','structure_quality']]
USA_project_ytrain = pd.DataFrame(train['SALE_PRC'])
```

```python
USA_project_xtrain
```

```python
USA_project_ytrain
```

```python
USA_project_xtest=test.loc[:,['LND_SQFOOT','TOT_LVG_AREA','SPEC_FEAT_VAL','RAIL_DIST','OCEAN_DIST','CNTR_DIST','SUBCNTR_DI','HWY_DIST','age','month_sold','structure_quality']]
USA_project_ytest = pd.DataFrame(test['SALE_PRC'])
```

```python
USA_project_xtest
```

```python
USA_project_ytest
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(USA_project_xtrain, USA_project_ytrain)
LinearRegression()
```

```python
USA_project_y_predict = model.predict(USA_project_xtest)
```

```python
from sklearn import metrics
import numpy as np
print('Miami_Housing_USA_project Mean Squared Error: ', metrics.mean_squared_error(USA_project_ytest, USA_project_y_predict))
print('Miami_Housing_USA_project Mean Absolute Error: ', metrics.mean_absolute_error(USA_project_ytest, USA_project_y_predict))
print('Miami_Housing_USA_project Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(USA_project_ytest, USA_project_y_predict)))
```

```python
#Support Vector Regression for Miami_Housing_USAproject
from sklearn.svm import SVR
model = SVR(kernel = 'linear')
model.fit(USA_project_xtrain, USA_project_ytrain)
```

```python
USA_project_y_predict=model.predict(USA_project_xtest)
```
